```python
fig,axes = plt.subplots(2,2,sharex=True,sharey=True)
for i in range(2):
    for j in range(2):
        axes[i,j].hist(randn(500),bins=50,color='k',alpha=0.5)
plt.subplots_adjust(wspace=0.3,hspace=0.3)  # 利用subplots_adjust()调整图像间的间距 ！！matplotlib不检查标签重叠
```

### plot函数可以接受一组x,y坐标，一个颜色变量和一个线型变量

```python
plt.plot(0,0,linestyle='--',color='g')
```

```python
plot(randn(30).cumsum(),color='g',linestyle='dashed',marker='o')  #marker强调实际数据点
```

## 设置标题、轴标签、刻度以及刻度标签


```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.plot(randn(1000).cumsum(),'k',label='one')
```

```python
ticks = ax.set_xticks([0,250,500,750,1000])    #将刻度放在数据范围那些位置
```

```python
labels = ax.set_xticklabels(['one','tow','three','four','five'],rotation=30,fontsize = 'small') #将刻度替换为指定标签
```

```python
ax.set_title('My Test')   #设置标题
ax.set_xlabel('Stages')   #设置X轴名称
```

```python
fig
```

## 添加图例(legend)


```python
ax.plot(randn(1000).cumsum(),'g--',label = 'two')
```

```python
ax.plot(randn(1000).cumsum(),'k.',label = 'three')
```

```python
ax.legend(loc = 'best') #loc 设定图例放置位置
```
